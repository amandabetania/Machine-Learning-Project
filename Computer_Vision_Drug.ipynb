```python
import pytesseract
from PIL import Image
from gtts import gTTS
from IPython.display import Audio, display
import re
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
# Extract text from the uploaded image using Tesseract OCR
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image, lang='eng')  # You can also use 'ind' for Indonesian
    return text.strip()

# Read text using gTTS
def read_text_with_gtts(text, lang='id'):
    tts = gTTS(text=text, lang=lang)
    tts.save("output_audio.mp3")
    return "output_audio.mp3"

# Test: replace with the name of the file you uploaded
image_file = list(uploaded.keys())[0]  # get the first uploaded file name
text_result = extract_text_from_image(image_file)
```

```python
def clean_text(raw_text):
    import re

    allowed_short_words = {'isi', 'dan', 'atau', 'untuk', 'yang'}  # valid short words to keep
    cleaned_lines = []

    for line in raw_text.split('\n'):
        line = line.strip()

        # Delete empty lines or lines with only 1 character
        if not line or len(line) < 2:
            continue

        # Delete common design symbols
        if re.match(r'^[^a-zA-Z0-9]+$', line):
            continue

        # Delete lines with random letter patterns like 'Vv' or 'Oo'
        if re.match(r'^[A-Za-z]{1,2}$', line) and line.lower() not in allowed_short_words:
            continue

        # Clean strange characters (except letters, numbers, and common punctuation)
        line = re.sub(r'[^\w\s.,;!?()\'\"-]', '', line)

        cleaned_lines.append(line)

    return '\n'.join(cleaned_lines).strip()
```

```python
# Extract, clean, and display the text
raw_text = extract_text_from_image(image_file)
cleaned_text = clean_text(raw_text)

# Display the cleaned result
print("üßπ Teks setelah dibersihkan:")
print(cleaned_text)
```

```python
# Convert cleaned text to speech
audio_path = read_text_with_gtts(cleaned_text, lang='id')
```

```python
display(Audio(audio_path, autoplay=True))
```
